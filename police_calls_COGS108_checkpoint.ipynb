```python
import pandas as pd
import numpy as np
import plotly.express as px
```

```python
police_calls_2023 = pd.read_csv('Data/pd_calls_for_service_2023_datasd.csv')
police_calls_2022 = pd.read_csv('Data/pd_calls_for_service_2022_datasd.csv')
police_calls_2021 = pd.read_csv('Data/pd_calls_for_service_2021_datasd.csv')
police_calls_2020 = pd.read_csv('Data/pd_calls_for_service_2020_datasd.csv')
```

```python
police_calls = pd.concat([police_calls_2023, police_calls_2022, police_calls_2021, police_calls_2020])
disp_codes = pd.read_csv('Data/pd_dispo_codes_datasd.csv')
beats = pd.read_csv('Data/pd_beat_codes_list_datasd.csv')
police_calls = police_calls.merge(disp_codes, left_on= 'disposition', right_on= 'dispo_code')
police_calls = police_calls.merge(beats)
police_calls = police_calls[['date_time', 'day_of_week', 'call_type', 'disposition', 'priority', 'description', 'neighborhood']]
police_calls = police_calls.sort_values(by = 'priority', ascending = False)
```
